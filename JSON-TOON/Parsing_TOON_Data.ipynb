```python
# How to do parsing TOON Data

import re

def parse(text):
    data = {}
    pattern = r'(\w+):\s*(.*)'

    for line in text.strip().splitlines():
        m=re.match(pattern, line)
        if not m:
            continue
        key, val = m.groups()

        val=val.strip()  # Remove the space

        #remove the quotes

        if val.startswith('"') and val.endswith('"'):
            val=val[1:-1]   # "Abinash Kumar" -> Abinash Kumar
        elif val.lower() in ("true","false"):
            val=val.lower() == "true"
        else:
            try:
                if "." in val:
                    val=float(val)
                else:
                    val=int(val)
            except:
                pass
            
        data[key]=val

    return data

toon_data="""
name:"Abinash Kumar
age:32
email: "abinash@gmail.com"
active: true"""

data=parse(toon_data)

print(f"Name: {data['name']}")
print(f"Age: {data['age']}")
print(f"Email: {data['email']}")
print(f"Active: {data['active']}")
```
