```python
#4  Generating TOON Data to JSON Data 

def toon_parse(text: str):
    lines = [l.strip() for l in text.strip().splitlines() if l.strip()]
    
    header = lines[0]
    name_part, rest = header.split("[", 1)
    name = name_part.strip()
    
    count_part, fields_part = rest.split("]", 1)
    fields_part = fields_part.strip()
    
    fields = fields_part[fields_part.find("{")+1 : fields_part.find("}")].split(",")
    fields = [f.strip() for f in fields]
    
    items = []
    for line in lines[1:]:
        values = [v.strip() for v in line.split(",")]
        row = {}
        for key, value in zip(fields, values):
            # auto convert int/float
            if value.replace(".", "", 1).isdigit():
                if "." in value:
                    row[key] = float(value)
                else:
                    row[key] = int(value)
            else:
                row[key] = value
        items.append(row)
    
    return {name: items}


toon_data = """
products[3]{id,name,price,stock}:
  101,Laptop,999.99,15
  102,Mouse,29.99,50
  103,Keyboard,79.99,30
"""

data = toon_parse(toon_data)

for product in data["products"]:
    print(product)
```
