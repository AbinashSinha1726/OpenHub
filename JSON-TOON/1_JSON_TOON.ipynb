```python
import json

def encode(records):
    if not records:
        return ""
    
    keys=list(records[0].keys())
    header=",".join(keys)

    rows=[]
    for r in records:
        rows.append(",".join(str(r[k]) for k in keys))
    return header + "\n" + "\n".join(rows)

with open("Sample.json","r", encoding="utf-8") as f:
    data=json.load(f)

    user_data=data["users"]
    print("JSON USer Data =", user_data)

    toon_data=encode(user_data)
    print("\n Toon USer Data =", toon_data)
```
