```python
import json
from tiktoken import get_encoding

enc=get_encoding("cl100k_base")

def count_tokens(text: str) -> int:
    return len(enc.encode(text))

def encode(records):
    if not records:
        return ""
    
    keys=list(records[0].keys())
    header=",".join(keys)

    rows=[]
    for r in records:
        rows.append(",".join(str(r[k]) for k in keys))
    return header + "\n" + "\n".join(rows)

with open("Sample.json","r", encoding="utf-8") as f:
    data=json.load(f)

    user_data=data["users"]
    print("JSON USer Data =", user_data)

    toon_data=encode(user_data)
    print("\n Toon USer Data =", toon_data)

    json_text=json.dumps(user_data,ensure_ascii=False)
    json_tokens=count_tokens(json_text)

    toon_tokens=count_tokens(toon_data)

    saved=json_tokens - toon_tokens
    percentage=(saved / json_tokens) * 100

    print("\n JSON Tokens: ",json_tokens)
    print("\n TOON Tokens: ",toon_tokens)
    print(f"\n Saved: {saved} tokens is ({percentage:.1f}%)")
```
