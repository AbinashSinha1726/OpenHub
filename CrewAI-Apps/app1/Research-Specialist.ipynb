# Conducts research on the Given topic and returns concise, verified insights.

```python
from dotenv import load_dotenv
import os

load_dotenv()

openai_api_key=os.getenv("OPENAI_API_KEY")
print("API Key is Loaded: ", openai_api_key is not None)
```

```python
from crewai.llm import LLM
llm=LLM(
    model="gpt-5.1",
    temperature=0.1
)
#response=llm.call("What is Crew AI?")
#print(response)
```

```python
from crewai import Agent, Task, Crew

research_agent=Agent(
    role="Research Specialist",
    goal="Research and compile the most important and recent findings related to the {topic}.",
    backstory="You are highly skilled at retrieving factual and contextually relevent insights.",
    verbose=True,
    llm=llm
)

writer_agent=Agent(
    role="Writer Specialist",
    goal="Craft a short and infomative long summary dervied from the research findings.",
    backstory="You have strong expertise in transforming source materials into concise summaries",
    verbose=True,
    llm=llm
)

task1 = Task(
    description="Provide 2-3 verified and timely facts about {topic}.",
    expected_output="Provide a list containing key facts.",
    agent=research_agent,
)

task2 = Task(
    description="Write a blog of 200 words summary about {topic} using the data from the research.",
    expected_output="Blog Summary",
    agent=writer_agent,
    context=[task1],
)

crew=Crew(
    agents=[research_agent,writer_agent],
    tasks=[task1,task2],
    verbose=True,
    tracing=True
)

result=crew.kickoff(inputs={"topic": "The future of Generative AI"})
print(result)
```
